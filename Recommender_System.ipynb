```python
# source'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#target ["Action","Adventure","Fantasy",SciFi]
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    
    lists=[]
    
    for i in ast.literal_eval(obj):
        lists.append(i["name"])
        
    return lists    
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
# same function will also be applied for kewords
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
#for cast
def convert3(obj):
    
    lists=[]
    counter=0
    
    for i in ast.literal_eval(obj):
        if counter==3:
            break
        lists.append(i["name"])
        counter+=1
        
        
    return lists    
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
# now from crew i only want to extract directors name
movies["crew"][0]
```

```python
def fetch_director(obj):
    
    L=list()
    counter=0
    
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
            break
        
        
    return L    
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

```python
movies["overview"][0]
```

```python
#converting the overview from string to list so that we could concatenate
movies["overview"]=movies["overview"].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# now remove the spaces between the words depicting the same thing
# for eg convert "Sam Worthington" to "SamWorthington" so as to make
#training easy
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
new_df=movies[["movie_id","title","tags"]]
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x:x.lower())
```

```python
#given a input movie our task is to give user the 5 movies which are almost similar to the input movie
# so we perform vectorization on tags using Bag of words techinque
#once vectorization is performed now it becomes easy for us to comapre the similarities between the tags
#what is the size of the vector would depend upon you

#here we would use sklearn

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()
```

```python
#but their are similar words like activity and activities as two different words here
#so to get rid of these we would use a technique called stemming

#for this we use nltk  library

!pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem("loving")
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
new_df["tags"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
vectors=cv.fit_transform(new_df["tags"]).toarray()
cv.get_feature_names()
```

```python
# now to find out the distance of each datapoint from the rest we use
#cosine similarity of sklearn

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
#this function returns the five most similar movies to the movie movie
def recommend(movie):
    
    movie_index=new_df[new_df["title"]==movie].index[0]
    
    distance=list(enumerate(similarity[movie_index]))
    
    sorted_distance=sorted(distance,reverse=True,key=lambda x:x[1])[1:6]
    
    for i in sorted_distance:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend("Batman Begins")
```
